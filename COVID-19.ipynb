# Covid-19 Data Analysis and Data Modeling
### Covid-19 Cases in INDIA From 2020-01-02 - 2020-12-11
1. In this project we are using Covid-19 Data set of India from 2020-01-02 - 2020-12-11 to perform analysis and data visulaization.
2. The data sets has fields are:
    -> Date
    -> Time
    -> State/UnionTerritory
    -> ConfirmedIndianNational 	
    -> ConfirmedForeignNational 	
    -> Cured
    -> Deaths
    -> Confirmed
3. This data set is taken from url: https://www.kaggle.com/sudalairajkumar/covid19-in-india

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid_19_india.csv', parse_dates=['Date'])
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'State/UnionTerritory':'State'}, inplace=True)
```

```python
df.head(2)
```

```python
df.groupby('State').sum()
```

```python
df.State = df.State.str.rstrip('*')
```

```python
df.groupby('State').sum()
```

```python
df.groupby('State')['ConfirmedIndianNational'].sum()
```

```python
df.ConfirmedIndianNational = df.ConfirmedIndianNational.apply(lambda x: np.nan if x == '-' else x)
```

```python
df.ConfirmedForeignNational = df.ConfirmedIndianNational.apply(lambda x: np.nan if x == '-' else x)
```

```python
df.isnull().sum()
```

```python
df.drop(['ConfirmedIndianNational', 'ConfirmedForeignNational'], axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.reset_index(inplace=True)
df.set_index('State', inplace=True)
```

```python
df.drop(['Sno'], axis=1, inplace=True)
```

```python
df.drop(['index'], axis=1, inplace=True)
```

```python
df.head(2)
```

### Latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) State wise.

```python
confirmed= df[['Confirmed']].groupby('State').sum().sort_values(by='Confirmed', ascending=False)
deaths= df[['Deaths']].groupby('State').sum().sort_values(by='Deaths', ascending=False)
recovered= df[['Cured']].groupby('State').sum().sort_values(by='Cured', ascending=False)
```

```python
confirmed.head()
```

```python
deaths.head()
```

```python
recovered.head()
```

```python
df['Active']= df.Confirmed-(df.Cured+df.Deaths) 
```

```python
df.head()
```

```python
active= df[['Active']].groupby('State').sum().sort_values(by= 'Active', ascending=False)
```

```python
active.head()
```

### Average Cases per month in each States of Novel Cornovirus(COVID-19).

```python
df1 = df[['Confirmed', 'Cured', 'Deaths', 'Active']].groupby([df.Date.dt.month, 'State']).sum().reset_index()
df1.set_index('Date', inplace=True)
```

### Visualize Confirmed, Deaths, Recovered and Active Cases of Novel Coronavirus(COVID-19) in Maharashtra in 12 Months Period.

```python
r_data = df1[df1.State == 'Maharashtra']

fig, ax = plt.subplots()

conf, = ax.plot(r_data.index, r_data.Confirmed, color='Blue')
conf.set_label('Confirmed')
death, = ax.plot(r_data.index, r_data.Deaths, color='Red')
death.set_label('Deaths')
recov, = ax.plot(r_data.index, r_data.Cured, color='Green')
recov.set_label('Recovered')
active, = ax.plot(r_data.index, r_data.Active, color='Yellow')
active.set_label('Active')

plt.title('Cases in Maharashtra in each Months', fontdict={'fontsize':15, 'fontweight':8})
plt.xlabel('Months', fontdict={'fontsize':15, 'fontweight':6})
plt.ylabel('Number of Cases\n(in Crore)', fontdict={'fontsize':15, 'fontweight':6})
plt.xticks(np.arange(0,13))

plt.legend()
plt.show()
```

### Calculate and Visualize Average of Confirmed, Deaths, Cured and Active Cases of Novel Coronavirus(COVID-19).

```python
data['AverageConfirmed'] = round(data.Confirmed/12)
data['AverageDeaths'] = round(data.Deaths/12)
data['AverageCured'] = round(data.Cured/12)
data['AverageActive'] = round(data.Active/12)
```

```python
data.head()
```

```python
r_data = data.sort_values(by='Confirmed', ascending=False).head().reset_index()
r_data = r_data[['State','AverageConfirmed','AverageDeaths','AverageCured','AverageActive']]
```

```python
r_data
```

```python
plt.figure(figsize=[15,10])
plt.subplots_adjust(wspace=0.4, hspace=0.4)

plt.subplot(2,2,1)
plt.barh(r_data.State, r_data.AverageConfirmed, color='dodgerblue')
plt.title('Average Confirmed Cases', fontdict={'fontsize':15,'fontweight':8})
plt.xlabel('Average Cases\n(Crore)', fontdict={'fontsize':15,'fontweight':6})
plt.ylabel('States', fontdict={'fontsize':15,'fontweight':6})

plt.subplot(2,2,2)
plt.barh(r_data.State, r_data.AverageDeaths, color='red')
plt.title('Average Deaths Cases', fontdict={'fontsize':15,'fontweight':8})
plt.xlabel('Average Cases', fontdict={'fontsize':15,'fontweight':6})
plt.ylabel('States', fontdict={'fontsize':15,'fontweight':6})

plt.subplot(2,2,3)
plt.barh(r_data.State, r_data.AverageCured, color='Green')
plt.title('Average Recovered Cases', fontdict={'fontsize':15,'fontweight':8})
plt.xlabel('Average Cases\n(Crore)', fontdict={'fontsize':15,'fontweight':6})
plt.ylabel('States', fontdict={'fontsize':15,'fontweight':6})

plt.subplot(2,2,4)
plt.barh(r_data.State, r_data.AverageActive, color='Yellow')
plt.title('Average Active Cases', fontdict={'fontsize':15,'fontweight':8})
plt.xlabel('Average Cases\n(Crore)', fontdict={'fontsize':15,'fontweight':6})
plt.ylabel('States', fontdict={'fontsize':15,'fontweight':6})


plt.show()
```
